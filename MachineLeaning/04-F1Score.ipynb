```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import GridSearchCV
import collections
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
import datetime
from sklearn.ensemble import RandomForestClassifier
```

# １、データ読み込むと前処理

```python
# データを読み込む
data = pd.read_csv("./x_train.csv", names=["col"])

# 説明変数を分離
data["col1"] = data["col"].map(lambda x:x.split(" ")[0])
data["col2"] = data["col"].map(lambda x:x.split(" ")[1])
data["col3"] = data["col"].map(lambda x:x.split(" ")[2])
data["col4"] = data["col"].map(lambda x:x.split(" ")[3])
data["col5"] = data["col"].map(lambda x:x.split(" ")[4])
data["col6"] = data["col"].map(lambda x:x.split(" ")[5])
data["col7"] = data["col"].map(lambda x:x.split(" ")[6])
data["col8"] = data["col"].map(lambda x:x.split(" ")[7])
data["col9"] = data["col"].map(lambda x:x.split(" ")[8])
data["col10"] = data["col"].map(lambda x:x.split(" ")[9])
data["col11"] = data["col"].map(lambda x:x.split(" ")[10])
data["col12"] = data["col"].map(lambda x:x.split(" ")[11])
data["col13"] = data["col"].map(lambda x:x.split(" ")[12])
data["col14"] = data["col"].map(lambda x:x.split(" ")[13])
data["col15"] = data["col"].map(lambda x:x.split(" ")[14])
data["col16"] = data["col"].map(lambda x:x.split(" ")[15])
data["col17"] = data["col"].map(lambda x:x.split(" ")[16])
data["col18"] = data["col"].map(lambda x:x.split(" ")[17])
```

# ２、訓練データ生成

```python
# 目的変数、説明変数を生成
X = data.iloc[:,1:]

y = pd.read_csv("./y_train.csv", names=["col"])
y = y.values.ravel()

print(X.shape, y.shape)
```

```python
X.head()
```

```python
# 標準化
stand = StandardScaler()
X_stand = stand.fit_transform(X)
```

```python
# 訓練データ生成
X_train, X_test, y_train, y_test = train_test_split(X_stand, y, test_size=0.3, random_state=666)
```

# ３、モデル構成

```python
# モデルで使用されたパラメータを設定する
parameter_random = {
    "n_estimators":[500],
    #"criterion":["gini", "entropy"] 
    #"max_features": ["sqrt", "log2"] 
    #"max_leaf_nodes":[24, 32]  0.32
}
```

```python
# ランダムフォレストをインスタンスする
random = RandomForestClassifier(bootstrap=True, oob_score=True)

# グリッドサーチを使用する
grid_svc = GridSearchCV(random, param_grid=parameter_random, scoring="f1", n_jobs = -1)
```

# ４、訓練と評価

```python
starttime = datetime.datetime.now()
grid_svc.fit(X_train, y_train)

endtime = datetime.datetime.now()
print((endtime - starttime).seconds) #20万 662
```

```python
#0.37962390257403
grid_svc.best_score_
```

```python
print("f1_score      :", f1_score(y_test,grid_svc.predict(X_test)))
print("accuracy_score:", accuracy_score(y_test,grid_svc.predict(X_test)))
print("recall_score  :", recall_score(y_test,grid_svc.predict(X_test)))
```
