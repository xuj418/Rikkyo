# 課題5
* 関数$f(x_1,x_2)=x_1^2+x_2^2$の最小値と、$f(x_1,x_2)$がその最小値をとるときの$x_1$と$x_2$の値を、PyTorchの自動微分を使って求めよう。

```python
import torch 
```

```python
# 関数の定義
def f(x1, x2, a=1.0, b=1.0):
  return a * x1 ** 2 + b * x2 ** 2
```

```python
# テンソルの準備
x1 = torch.tensor(10.0, requires_grad=True)
x2 = torch.tensor(20.0, requires_grad=True)

# 最適化手法のインスタンスを作る
#   param: 変数で微分する
#   lr: 学習率
optimizer = torch.optim.SGD(params=[x1, x2], lr=0.1)
```

```python
for i in range(1, 101):
    
    # 最適化対象のテンソルの勾配を0で初期化します。
    # backward()を呼び出す前にzero_grad()を行う。
    optimizer.zero_grad()

    y = f(x1, x2)
    # 誤差逆伝播
    y.backward()

    # 学習、パラメータを更新
    # backward()を呼び出す後にstep()を行う
    optimizer.step()
    
    if i % 5 == 0:
        print(f'iter {i} : f(x) = {y.data:.6f}, x1 = {x1.data:.6f}, x2 = {x2.data:.6f}')
```
